## Q4 Find the optimum value of k for k-means clustering using the elbow method. Plot the elbow curve

```python
from sklearn.cluster import KMeans
n_clusters=15
cost=[]
for i in range(1,n_clusters):
    kmean= KMeans(i)
    kmean.fit(df)
    cost.append(kmean.inertia_)
plt.plot(cost, 'bx-')    
```

## Q5 Find the optimum value of k for k-means clustering using the silhouette score method and specify the number of observations in each cluster using a bar plot

```python
import sklearn.metrics as metrics
for i in range(2,6):
    kmeans=KMeans(n_clusters=i,init="k-means++",random_state=200)
    kmeans.fit(df)
    labels=kmeans.labels_
    print("Silhouette  score for k(clusters)="+str(i)+" is "+str(metrics.silhouette_score(df,labels,metric="euclidean",sample_size=1000,random_state=200)))
```
